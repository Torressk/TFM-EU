```python
import pandas as pd

# Lee el archivo CSV
file_path = '/content/stats-2.csv'
df = pd.read_csv(file_path, sep=',')

# Muestra las primeras filas del DataFrame
print(df.head())
```

```python
def clean_data(file_path, sep=','):
    df = pd.read_csv(file_path, sep=sep)
    df.fillna(0, inplace=True)

    # Asegúrate de que las columnas existen y contienen el valor esperado antes de aplicar split
    if 'Player' in df.columns:
        df['Player'] = df['Player'].apply(lambda x: x.split('\\')[0] if '\\' in x else x)
    if 'Pos' in df.columns:
        df['Pos'] = df['Pos'].apply(lambda x: x[:2] if isinstance(x, str) else x)
    if 'Nation' in df.columns:
        df['Nation'] = df['Nation'].apply(lambda x: x.split('\\')[1] if '\\' in x else x)

    # Si 'Matches' no es una columna en tu DataFrame, este paso lanzará un error
    if 'Matches' in df.columns:
        df.drop('Matches', axis=1, inplace=True)

    return df

# Aplica la función de limpieza
df_premierleague_23 = clean_data(file_path)
print(df_premierleague_23.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

# Diccionario con los jugadores y sus imágenes
Imagenes_players = {
    'Erling Haaland': '/content/Haland.png',
    'Cole Palmer': '/content/Cole Palmer.png',
    'Alexander Isak': '/content/Alexander Isak.png',
    'Phil Foden': '/content/Foden.png',
    'Dominic Solanke': '/content/Dominic.png'
}

# Crear el DataFrame de los máximos goleadores (asegúrate de que df_top_5_goleadores esté definido correctamente)
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(data=df_top_5_goleadores, x='Gls', y='xG', alpha=0.7)

zoom = 0.2

for Player, row in df_top_5_goleadores.iterrows():
    if Player in Imagenes_players:
        x = row['Gls']
        y = row['xG']
        print(f"Player: {Player}, Coords: ({x}, {y})")

        img = plt.imread(Imagenes_players[Player])
        img = OffsetImage(img, zoom=zoom)
        ab = AnnotationBbox(img, (x, y), frameon=False)
        scatter.add_artist(ab)

plt.title('Grafico comparativo del top 5 de maximos goleadores')
plt.xlabel('Goles')
plt.ylabel('xG')
plt.grid(True)
plt.show()
```
